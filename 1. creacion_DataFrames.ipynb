**tabla principal // DF_OFERTAS**

```python
#DF_OFERTAS

map_empresa = {v : k for k, v in enumerate(df_empresa.iloc[:,0], start=1)}
map_ubicacion = {v : k for k, v in enumerate(df_ubicacion.iloc[:,0], start=1)}
map_modalidad = {v : k for k, v in enumerate(df_modalidad.iloc[:,0], start=1)}
map_jornada = {v : k for k, v in enumerate(df_jornada.iloc[:,0], start=1)}
map_contrato = {v : k for k, v in enumerate(df_contrato.iloc[:,0], start=1)}
map_portal = {v : k for k, v in enumerate(df_portal_empleo.iloc[:,0], start=1)}

df_ofertas = ofertas_empleo[["id_urls",
                            "fecha",
                            "oferta",
                            "empresa",
                            "funcion",
                            "salario_min",
                            "salario_max",
                            "estudios",
                            "experiencia",
                            "ubicacion",
                            "modalidad",
                            "jornada",
                            "contrato",
                            "skills",
                            "tecnologias_aptitudes",
                            "idiomas",
                            "vacaciones",
                            "beneficios",
                            "portal_empleo",
                            "descripcion"]]


df_ofertas["empresa"] = df_ofertas["empresa"].map(map_empresa)
df_ofertas["ubicacion"] = df_ofertas["ubicacion"].map(map_ubicacion)
df_ofertas["modalidad"] = df_ofertas["modalidad"].map(map_modalidad)
df_ofertas["jornada"] = df_ofertas["jornada"].map(map_jornada)
df_ofertas["contrato"] = df_ofertas["contrato"].map(map_contrato)
df_ofertas["portal_empleo"] = df_ofertas["portal_empleo"].map(map_portal)


df_ofertas = df_ofertas.rename(columns={"empresa" : "id_empresa",
                                        "ubicacion" : "id_ubicacion",
                                        "modalidad" : "id_modalidad",
                                        "jornada" : "id_jornada",
                                        "contrato" : "id_contrato",
                                        "portal_empleo" : "id_portal_empleo"})


df_ofertas = df_ofertas.replace({np.nan : None})

columnas_id = ["id_empresa", "id_ubicacion", "id_modalidad", "id_jornada", "id_contrato", "id_portal_empleo"]

for col in columnas_id:
    df_ofertas[col] = df_ofertas[col].apply(lambda x: int(x) if x is not None else None)

df_ofertas
```

**skills, tecnologias_aptitudes, idiomas, beneficios // (VARIOS_VALORES)**

```python
#DF_SKILLS

skills_list = ofertas_empleo["skills"].dropna().apply(lambda x: x.split(',')).sum()

df_skills = pd.DataFrame(set([skill.strip() for skill in skills_list]), columns=["skills"])

df_skills = df_skills.reset_index(drop=True)

df_skills["id_skills"] = df_skills.index + 1

df_skills
```

```python
#DF_TECNOLOGIAS_APTITUDES

tecnologias_aptitudes_list = ofertas_empleo["tecnologias_aptitudes"].dropna().apply(lambda x: x.split(',')).sum()

df_tecnologias_aptitudes = pd.DataFrame(set([tecnologias_aptitudes.strip() for tecnologias_aptitudes in tecnologias_aptitudes_list]), columns=["tecnologias_aptitudes"])

df_tecnologias_aptitudes = df_tecnologias_aptitudes.reset_index(drop=True)

df_tecnologias_aptitudes["id_tecnologias"] = df_tecnologias_aptitudes.index + 1

df_tecnologias_aptitudes
```

```python
#DF_IDIOMAS

idiomas_list = ofertas_empleo["idiomas"].dropna().apply(lambda x: x.split(',')).sum()

df_idiomas = pd.DataFrame(set([idiomas.strip() for idiomas in idiomas_list]), columns=["idiomas"])

df_idiomas = df_idiomas.reset_index(drop=True)

df_idiomas["id_idiomas"] = df_idiomas.index + 1

df_idiomas
```

```python
#DF_BENEFICIOS

beneficios_list = ofertas_empleo["beneficios"].dropna().apply(lambda x: x.split(',')).sum()

df_beneficios = pd.DataFrame(set([beneficios.strip() for beneficios in beneficios_list]), columns=["beneficios"])

df_beneficios = df_beneficios.reset_index(drop=True)

df_beneficios["id_beneficios"] = df_beneficios.index + 1

df_beneficios
```

**oferta_tecnologia, oferta_skill, oferta_idioma, oferta_beneficio // (COMPUESTAS)**

```python
#DF_OFERTA_TECNOLOGIA

map_tecnologias = dict(zip(df_tecnologias_aptitudes["tecnologias_aptitudes"], df_tecnologias_aptitudes["id_tecnologias"]))

df_oferta_tecnologia = ofertas_empleo[["id_urls", "tecnologias_aptitudes"]].dropna()

df_oferta_tecnologia["tecnologias_aptitudes"] = df_oferta_tecnologia["tecnologias_aptitudes"].apply(
    lambda x: [tech.strip() for tech in x.split(',')]
)

df_oferta_tecnologia = df_oferta_tecnologia.explode("tecnologias_aptitudes").reset_index(drop=True)

df_oferta_tecnologia["tecnologias_aptitudes"] = df_oferta_tecnologia["tecnologias_aptitudes"].map(map_tecnologias)

df_oferta_tecnologia.columns = ["id_urls", "id_tecnologias"]

df_oferta_tecnologia = df_oferta_tecnologia.dropna().reset_index(drop=True)

df_oferta_tecnologia["id_tecnologias"] = df_oferta_tecnologia["id_tecnologias"].astype(int)

df_oferta_tecnologia
```

```python
#DF_OFERTA_SKILL

map_skill = dict(zip(df_skills["skills"], df_skills["id_skills"]))

df_oferta_skill = ofertas_empleo[["id_urls", "skills"]].dropna()

df_oferta_skill["skills"] = df_oferta_skill["skills"].apply(lambda x: [skill.strip() for skill in x.split(',')])

df_oferta_skill = df_oferta_skill.explode("skills").reset_index(drop=True)

df_oferta_skill["skills"] = df_oferta_skill["skills"].map(map_skill)

df_oferta_skill.columns = ["id_urls", "id_skills"]

df_oferta_skill = df_oferta_skill.dropna().reset_index(drop=True)

df_oferta_skill["id_skills"] = df_oferta_skill["id_skills"].astype(int)

df_oferta_skill
```

```python
#DF_OFERTA_IDIOMA

map_idioma = dict(zip(df_idiomas["idiomas"], df_idiomas["id_idiomas"]))

df_oferta_idioma = ofertas_empleo[["id_urls", "idiomas"]].dropna()

df_oferta_idioma["idiomas"] = df_oferta_idioma["idiomas"].apply(lambda x: [idioma.strip() for idioma in x.split(',')])

df_oferta_idioma = df_oferta_idioma.explode("idiomas").reset_index(drop=True)

df_oferta_idioma["idiomas"] = df_oferta_idioma["idiomas"].map(map_idioma)

df_oferta_idioma.columns = ["id_urls", "id_idiomas"]

df_oferta_idioma = df_oferta_idioma.dropna().reset_index(drop=True)

df_oferta_idioma["id_idiomas"] = df_oferta_idioma["id_idiomas"].astype(int)

df_oferta_idioma
```

```python
#DF_OFERTA_BENEFICIO

map_beneficio = dict(zip(df_beneficios["beneficios"], df_beneficios["id_beneficios"]))

df_oferta_beneficio = ofertas_empleo[["id_urls", "beneficios"]].dropna()

df_oferta_beneficio["beneficios"] = df_oferta_beneficio["beneficios"].apply(lambda x: [benef.strip() for benef in x.split(',')])

df_oferta_beneficio = df_oferta_beneficio.explode("beneficios").reset_index(drop=True)

df_oferta_beneficio["beneficios"] = df_oferta_beneficio["beneficios"].map(map_beneficio)

df_oferta_beneficio.columns = ["id_urls", "id_beneficios"]

df_oferta_beneficio = df_oferta_beneficio.dropna().reset_index(drop=True)

df_oferta_beneficio["id_beneficios"] = df_oferta_beneficio["id_beneficios"].astype(int)

df_oferta_beneficio
```

```python
def insert_to_table(data, table, database, auto_increment, host = "localhost", user = "root", password = "contraseña1234"):
    
    db = mysql.connector.connect(host     = host,
                                 user     = user,
                                 password = password,
                                 database = database)
    cursor = db.cursor()

    cursor.execute(f"SELECT * FROM {table} LIMIT 0;")
    if auto_increment:
        if len(cursor.column_names) == 1:
            column_names = cursor.column_names
        else:
            column_names = cursor.column_names[1:]
    else:
        column_names = cursor.column_names
        
    cursor.fetchall()

    insert_query = f"INSERT INTO {table} ({', '.join(column_names)}) VALUES ({', '.join(['%s' for _ in column_names])})".replace("'", "")
    print(insert_query[:500])
    values = [tuple(row) for row in data]
    # print(values[:100])

    # .executemany ejecuta el query de INSERT INTO con cada uno de los elementos de "values"
    cursor.executemany(insert_query, values)
    
    # Guarda los resultados
    db.commit()

    print(f"Añadidas: {cursor.rowcount} filas")

    cursor.fetchall() # Vaciamos el cursor
    cursor.close()
    db.close()
```

```python
database = "ofertas_empleo"

#tablas CON autoincrement: empresa, ubicacion, modalidad, jornada, contrato, portal_empleo, skills, tecnologias_aptitudes, idiomas, beneficios
tablas_dataframes_auto = {"empresa": df_empresa,
                          "ubicacion": df_ubicacion,
                          "modalidad": df_modalidad,
                          "jornada": df_jornada,
                          "contrato": df_contrato,
                          "portal_empleo": df_portal_empleo        
                         }

#tablas SIN autoincrement:
tablas_dataframes = {
                    "tecnologias_aptitudes": df_tecnologias_aptitudes[["id_tecnologias", "tecnologias_aptitudes"]],
                    "idiomas": df_idiomas[["id_idiomas", "idiomas"]],
                    "beneficios": df_beneficios[["id_beneficios", "beneficios"]],
                    "skills": df_skills[["id_skills", "skills"]],
                    "ofertas": df_ofertas.drop(["skills", "tecnologias_aptitudes", "idiomas", "beneficios"], axis = 1).replace({np.nan : None}).drop_duplicates("id_urls"),
                    "oferta_tecnologia": df_oferta_tecnologia.drop_duplicates("id_urls"),
                    "oferta_skill": df_oferta_skill.drop_duplicates("id_urls"),
                    "oferta_idioma": df_oferta_idioma.drop_duplicates("id_urls"),
                    "oferta_beneficio": df_oferta_beneficio.drop_duplicates("id_urls")
                    }

#tablas CON autoincrement:
for tabla, df in tablas_dataframes_auto.items():
    insert_to_table(df.values, table=tabla, database=database, auto_increment = True)

#tablas SIN autoincrement:
for tabla, df in tablas_dataframes.items():
    insert_to_table(df.values, table=tabla, database=database, auto_increment = False)
```
