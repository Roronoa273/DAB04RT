```python
import pandas as pd
df_final = pd.read_csv("df_final.csv",
                      dtype={"salario_min": "float64",
                             "salario_max": "float64",
                             "estudios": "bool",
                             "experiencia": "float64",
                             "vacaciones": "Int64"},
                      parse_dates=["fecha", "fecha_extraccion"])
```

```python
df_final['descripcion'] = df_final['descripcion'].fillna('').astype(str)
df_final['skills'] = df_final['skills'].fillna('')
habilidades_unicas = (df_final['skills'].str.split(',').explode().str.strip().dropna().unique())
habilidades_unicas = [h for h in habilidades_unicas if h]

# 3. Función para completar habilidades desde descripción
def completar_habilidades(row):
    if row['skills'].strip() == '':
        habilidades_encontradas = [h for h in habilidades_unicas if h.lower() in row['descripcion'].lower()]
        return ', '.join(habilidades_encontradas) if habilidades_encontradas else None
    return row['skills']

# 4. Aplicar función
df_final['skills'] = df_final.apply(completar_habilidades, axis=1)
```
